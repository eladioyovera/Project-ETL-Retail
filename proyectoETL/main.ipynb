```python
from etl.extract import Extract
from etl.transform import Transform
from etl.load import Load
import pandas as pd
```

```python
extract = Extract()
transform = Transform()
load = Load()
```

```python
df_customer = extract.read_csv("../data/customer.csv",",",["customer_id","customer_fname","customer_lname","customer_email","customer_password","customer_street","customer_city","customer_state","customer_zipcode"])
df_orders = extract.read_json("../data/orders.json")
df_order_items = extract.read_json("../data/order_items.json")
df_departments = extract.read_csv("../data/departments.csv","~",['department_id', 'department_name'])
df_categories = extract.read_tsv("../data/categories.tsv",["category_id", "category_department_id", "category_name"])
df_products = extract.read_xml("../data/products.xml")
```
